```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train)
```

```python
from sklearn.preprocessing import LabelEncoder
le_x=LabelEncoder()
```

```python
for i in range(0,6):
    x_train[:,i]=le_x.fit_transform(x_train[:,i])
```

```python
x_train[:,7]=le_x.fit_transform(x_train[:,7])
```

```python
x_train
```

```python
le_y=LabelEncoder()
```

```python
y_train=le_y.fit_transform(y_train)
y_train
```

```python
for i in range(0,6):
    x_test[:,i]=le_x.fit_transform(x_test[:,i])
```

```python
x_test[:,7]=le_x.fit_transform(x_test[:,7])
```

```python
x_test
```

```python
y_test=le_y.fit_transform(y_test)
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x_train=ss.fit_transform(x_train)
x_test=ss.fit_transform(x_test)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC=DecisionTreeClassifier(criterion='entropy',random_state=0)
DTC.fit(x_train,y_train)
```

```python
y_pred=DTC.predict(x_test)
y_pred
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_pred,y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
nbc=GaussianNB()
nbc.fit(x_train,y_train)
```

```python
y_pred=nbc.predict(x_test)
y_pred
```

```python
print(metrics.accuracy_score(y_pred,y_test))
```

```python
from sklearn.svm import SVC
sv=SVC(kernel='linear',random_state=0)
sv.fit(x_train,y_train)
```

```python
y_pred=sv.predict(x_test)
y_pred
```

```python
print(metrics.accuracy_score(y_pred,y_test))
```

```python
sv=SVC(kernel='rbf',random_state=0)
sv.fit(x_train,y_train)
```

```python
y_pred=sv.predict(x_test)
y_pred
```

```python
print(metrics.accuracy_score(y_pred,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression(random_state=0)
lg.fit(x_train,y_train)
```
